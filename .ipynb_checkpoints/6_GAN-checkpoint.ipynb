```python
import torch, torchvision
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as opt
import torchvision.transforms as transforms
import numpy as np
from PIL import Image
import os, glob
os.environ['CUDA_VISIBLE_DEVICES'] = '0, 1'
from torch.autograd import Variable
from torchvision.datasets import ImageFolder
device = "cuda"
```

```python
class Mydatasets(torch.utils.data.Dataset):
    def __init__(self, folder, transform = None):
        self.transform = transform
        self.files = glob.glob(os.path.join(folder,'*.jpg'))
        self.images = []
        for fn in self.files:
            try:
                image = Image.open(fn)
                if self.transform:
                    image = self.transform(image)
                if image.shape[0] != 3:
                    continue
                    #gray画像ははじく
            except IOError:
                continue
            self.images.append(image)
    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        out_data = self.images[idx]
        return out_data
#データセット定義　指定されたfolderのjpgをすべて読み込む
```

```python

transform = transforms.Compose([
    transforms.Resize((64, 64)),
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])
#前処理は64*64へのリサイズ[-1,1]クリッピング,Tensorへの変換のみ

dataset = Mydatasets(folder="/export/space/yanai/media/foodimg64/ramen/", transform=transform)
#ラーメン画像を学習画像に使う
#dataset = torchvision.datasets.MNIST(root="dataset",download=True)v
#dataset = ImageFolder(root="/export/data/dataset/UECFOOD/UECFOOD100",transform=transform)
```

```python
class Generator(nn.Module):
    def __init__(self):
        super(Generator, self).__init__()        
        self.main = nn.Sequential(

            nn.ConvTranspose2d(100, 256, 4, 1, 0),
            nn.BatchNorm2d(256),
            nn.LeakyReLU(0.1),

            nn.ConvTranspose2d(256, 128, 4, 2, 1),
            nn.BatchNorm2d(128),
            nn.LeakyReLU(0.1),

            nn.ConvTranspose2d(128, 64, 4, 2, 1),
            nn.BatchNorm2d(64),
            nn.LeakyReLU(0.1),

            nn.ConvTranspose2d(64, 32, 4, 2, 1),
            nn.BatchNorm2d(32),
            nn.LeakyReLU(0.1),

            nn.ConvTranspose2d(32, 3, 4, 2, 1),
            nn.Tanh()
        )
    def forward(self, x):
        return self.main(x)
#(batch_size,100, 1, 1をdeconvして(batch_size, 3,64,64)の画像を生成
#メディア実験を参考に
```

```python
g = Generator()
t=Variable(torch.zeros((1,100,1,1)))
fake = g(t)
print(fake.shape)
```

```python
from torch.nn.utils.spectral_norm import spectral_norm
#メディア実験を参考にしたSN-GANを用いる
#https://pytorch.org/docs/master/generated/torch.nn.utils.spectral_norm.html
class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator, self).__init__()        
        self.main = nn.Sequential(
            spectral_norm(nn.Conv2d(3, 64, stride=2, kernel_size=3, padding=1)),
            nn.LeakyReLU(0.1),
            spectral_norm(nn.Conv2d(64, 128, stride=2, kernel_size=3, padding=1)),
            nn.LeakyReLU(0.1),
            spectral_norm(nn.Conv2d(128, 256, stride=2, kernel_size=3, padding=1)),
            nn.LeakyReLU(0.1),
            spectral_norm(nn.Conv2d(256, 512, stride=2, kernel_size=3, padding=1)),
            nn.LeakyReLU(0.1),
            spectral_norm(nn.Conv2d(512, 1, stride=1, kernel_size=3, padding=1)),
            nn.LeakyReLU(0.1),
            nn.AdaptiveAvgPool2d(output_size=(1,1)),
            nn.Flatten(),
            
        )
    def forward(self, x):
        return self.main(x).squeeze_(1)
#d=Discriminator()
#out = d(fake)
#lossfunc = nn.CrossEntropyLoss()
#lossfunc = nn.BCEWithLogitsLoss()
#lossfunc = nn.HingeEmbeddingLoss()
#loss = lossfunc(out, Variable(torch.zeros((1),dtype=torch.long)))
#loss.backward()
```

```python

from IPython import display
import matplotlib.pyplot as plt
%matplotlib inline
#メディア実験のプロット関数を少し変更
#乱数からr*c枚の画像を生成して並べて表示
def sample_imgs(generator):
    r, c = 4, 10
    imgs = generator(torch.randn((r*c,100,1,1)).to(device))
    fig, axs = plt.subplots(r, c, figsize=(16,6))
    for i in range(r):
        for j in range(c):
            axs[i,j].imshow(((
                imgs.detach()[i*c+j].cpu().numpy().transpose(1,2,0)+1)*128).astype("int16"))
            axs[i,j].axis('off')
    display.clear_output(wait = True)
    display.display(fig)
    plt.close()
```

```python

d=Discriminator()
g=Generator()
d = d.to(device)
g = g.to(device)
g_opt = opt.Adam(g.parameters(),lr=0.0002, betas=(0.5, 0.999))
d_opt = opt.Adam(d.parameters(),lr=0.0002, betas=(0.5, 0.999))
#OptimizerはともにAdamを用いる
batch_size=16
dataloader = torch.utils.data.DataLoader(dataset, batch_size=batch_size, shuffle=True)
#データローダにデータセットをセット
#a=torch.cat((aa,bb),axis=1)
#b=torch.cat((bb,aa),axis=1)
for ep in range(1000):
    for (real,i) in zip(dataloader,range(len(dataloader))):
        real = real.to(device)
        fake = g(torch.randn((real.shape[0],100,1,1)).to(device))
        #fake画像生成
        #print(d(fake).shape)
        
        #generatorの学習
        g_loss = - torch.mean(d(fake))
        g_opt.zero_grad()
        g_loss.backward()
        g_opt.step()
        
        #discriminatorの学習
        d_opt.zero_grad()
        #fake = g(Variable(torch.randn((batch_size,100,1,1))).to(device))
        d_loss = torch.mean(-torch.min(torch.zeros((real.shape[0])).to(device),-1 + d(real)) - torch.min(torch.zeros((real.shape[0])).to(device),-1-d(fake.detach())))
        #lossはhinge lossを用いる
        d_loss.backward()
        d_opt.step()
    if ep % 10 == 0:
        #生成画像,lossの確認
        sample_imgs(g)
        print("epoch: %d"%ep)
        print("discriminator loss: %f"%d_loss)
        print("generator loss: %f"%g_loss)
    
```

Spectral normalizationを用いると結果が見るからにうまくいくことが分かったがどうしてうまくいくのかはよくわからなかった。ノイズからの生成なので
生成される分布をデータセット画像に合わせるには時間がかかるのかと思った。また、用いるデータセットを多くすればもう少し良い結果が得られるのではないかと考える。
データセットの質も、あまりに外れ値が多いと学習がうまくいかないと感じた。
