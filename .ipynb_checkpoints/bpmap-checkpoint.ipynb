```python
import torch, torchvision
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as opt
import torchvision.transforms as transforms
import numpy as np
import os
from PIL import Image
from IPython import display
import matplotlib.pyplot as plt
%matplotlib inline
device = "cuda"
```

```python
model = torchvision.models.vgg16(pretrained=True)
model = model.to(device)
model.eval()
#モデルはVGG16を用いる
```

```python
f = open('../1000.txt')
text1000 = f.read().split("\n")
f.close()
#imagenet1000種類のテキストラベル
```

```python

preprocess = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])
#画像を224*224にし、平均と分散をimagenetに合わせる
```

```python
#普通のback propagation
def create_bpmap(img_fn):
    image = Image.open(img_fn)
    #引数のファイル名から画像読み込み
    display.display(image)
    #とりあえず表示
    image = preprocess(image).to(device)
    image = image.reshape(1,*image.shape)
    #前処理して(1,3,224,224)の形にする
    
    plt.imshow(np.array(image.cpu())[0].transpose(1,2,0))
    plt.show()
    #前処理が終わった画像を表示
    
    image.requires_grad = True
    #勾配を画像まで伝えるようにする
    
    output = model(image)
    idx = torch.argmax(output)
    #forwardして最も出力が高いものを認識ラベルとし、その反応マップを作る
    
    print(text1000[idx])
    #分類されたクラスラベル表示
    
    dzdy = torch.zeros(1,1000).to(device)
    dzdy[0][idx]=1
    output.backward(dzdy)
    #分類されたラベルのみ1のone hotベクタを作り、それをbackwardすることでそのクラスの出力が変化するところが勾配として現れる
    
    bpmap = image.grad[0]
    #画像の勾配を取り出す
    bpmap = (bpmap / torch.max(bpmap))
    #正規化
    print(bpmap.shape)
    plt.imshow(bpmap.cpu().numpy().transpose(1,2,0))
    plt.show()
    #反応マップの表示
```

```python
create_bpmap("/export/data/dataset/animal/dog/415000.jpg")
```

5枚の画像いずれも物体の位置に反応が強く出ていることがわかる。
猫の画像だと背景が全く関係ないので反応が背景にはほとんど出ていないが、
ほかの画像だとその生物がすむ環境などであり、その動物に関連する要素なため反応が出てしまっているのではないかと考える。

```python
#smooth grad
#ノイズを加えて10個back propergationし、その平均を取る
def create_smooth_bpmap(img_fn):
    image = Image.open(img_fn)
    #ファイル名から画像を開く
    
    display.display(image)
    #とりあえず表示
    
    image = preprocess(image).to(device)
    image = image.reshape(1,*image.shape)
    #画像を(1,3,224,224)に変形
    
    image = image.expand(10,*image.shape[1:])
    #画像を10個にコピーして(10,3,224,224)の形にする
    
    image = image + (torch.randn(*image.shape).to(device)*2e-1)
    #標準偏差0.02のノイズを足す
    
    plt.imshow(np.array(image.cpu())[0].transpose(1,2,0))
    plt.show()
    #変形してノイズ足した後の画像を1枚表示
    
    image.requires_grad = True
    #画像まで勾配を伝播させる
    
    output = model(image)
    idx = torch.argmax(output[0])
    #最も出力が高いものを認識ラベルとする
    print(text1000[idx])
    #認識ラベルのテキストを表示
    
    dzdy = torch.zeros(10,1000).to(device)
    dzdy[:,idx]=1
    output.backward(dzdy)    
    #分類されたラベルのみ1のone hotベクタを作り、それをbackwardすることでそのクラスの出力が変化するところが勾配として現れる
   
    
    bpmap = torch.mean(image.grad,axis=0)
    #10枚の画像の勾配の平均を取る
    
    bpmap = (bpmap / torch.max(bpmap))
    #正規化
    
    plt.imshow(bpmap.cpu().numpy().transpose(1,2,0))
    plt.show()
    #アクティベーションマップを表示
```

```python
create_smooth_bpmap("/export/data/dataset/animal/penguin/541049.jpg")
```

ペンギンの輪郭の部分によりはっきり反応が出ているのが確認できる。特に翼の部分は勾配1回のみの手法だとあまり出ていなかったのでわかりやすくなったと思った。

```python
#https://github.com/utkuozbulak/pytorch-cnn-visualizations/blob/master/src/guided_backprop.py
```

```python
guided_model = torchvision.models.vgg16(pretrained=True)
#VGG16のRelu層を書き換える
#forwardはそのまま用いるが、backwardの時に正の勾配のみ伝播するようにする
#https://github.com/utkuozbulak/pytorch-cnn-visualizations/blob/master/src/guided_backprop.py
#を参考にしている
relu_forward_output=[]
def relu_forward_hook(module, ten_in, ten_out):
    relu_forward_output.append(ten_out)
    #forward時のoutputを保持
def relu_backward_hook(module, grad_in, grad_out):
    forward_output = relu_forward_output[-1]
    forward_output[forward_output>0]=1
    #forward時のoutputを持ってきて、0以外の場所を1にした0か1の配列にする
    modified_grad_out = torch.clamp(grad_in[0], min=0.0)*forward_output
    #勾配の0以下を0に置き換え、reluのbackwardと同じくforward時に0にしたところを0にする
    
    del relu_forward_output[-1]
    return (modified_grad_out,)

for pos, module in guided_model.features._modules.items():
    if isinstance(module, nn.ReLU):
        module.register_backward_hook(relu_backward_hook)
        module.register_forward_hook(relu_forward_hook)
#モデルの各reluに順に適用

guided_model = guided_model.to(device)
```

```python
def create_guided_bpmap(img_fn):
    image = Image.open(img_fn)
    #ファイル名から画像読み込み
    display.display(image)
    image = preprocess(image).to(device)
    image = image.reshape(1,*image.shape)
    #画像を(1,3,224,224)に変形
    
    plt.imshow(np.array(image.cpu())[0].transpose(1,2,0))
    plt.show()
    #変形後画像表示
    
    image.requires_grad = True
    #勾配が画像まで伝播するように
    
    output = guided_model(image)
    idx = torch.argmax(output)
    #最も出力が高いものを認識ラベルとする
    print(text1000[idx])
    #認識ラベルのテキストを表示
    
    dzdy = torch.zeros(1,1000).to(device)
    dzdy[0][idx]=1
    output.backward(dzdy)    
    #分類されたラベルのみ1のone hotベクタを作り、それをbackwardすることでそのクラスの出力が変化するところが勾配として現れる

    
    bpmap = image.grad[0]
    #画像の勾配を取り出しアクティベーションマップとする
    
    bpmap = (bpmap / torch.max(bpmap))
    #正規化
    plt.imshow(bpmap.cpu().numpy().transpose(1,2,0))
    plt.show()
    #アクティベーションマップの表示
```

```python
create_guided_bpmap("/export/data/dataset/animal/penguin/541049.jpg")
```

```python
create_guided_bpmap("/export/data/dataset/animal/cat/458050.jpg")
```

単純な手法の割にはﾌﾂｳﾉBackpropagationと比べて明確に輪郭がとらえられているとわかる。
猫だとほかの部位より目が特に反応が強いのは面白いと感じた。
