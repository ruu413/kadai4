```python
import torch, torchvision
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as opt
import torchvision.transforms as transforms
import numpy as np
import os
from PIL import Image
from IPython import display
import matplotlib.pyplot as plt
import glob
%matplotlib inline
device = "cuda"
```

```python
preprocess = transforms.Compose([
    transforms.Resize((256,256)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5]),
])
#256*256にリサイズして[-1,1]にクリッピング
```

```python
class CycleGANDataset(torch.utils.data.Dataset):
    def __init__(self, folder):
        self.files_a = glob.glob(os.path.join(folder+'A/*.jpg'))
        self.files_b = glob.glob(os.path.join(folder+'B/*.jpg'))
        self.images = []
        for fn_a, fn_b in zip(self.files_a, self.files_b):
            if len(self.images) % 100 == 0:
                print(len(self.images))
            try:
                image_a = preprocess(Image.open(fn_a))
                image_b = preprocess(Image.open(fn_b))
            except IOError:
                continue
            self.images.append((image_a, image_b))
    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        out_data = self.images[idx]
        return out_data
#データセットはA,Bのタプルが一つのデータとする A,Bはペア画像ではない
dataset = CycleGANDataset("/home/yanai-lab/terauchi-k/export/jupyter/notebook/pytorch-CycleGAN-and-pix2pix/datasets/datasets/apple2orange/train")
dataset_test = CycleGANDataset("/home/yanai-lab/terauchi-k/export/jupyter/notebook/pytorch-CycleGAN-and-pix2pix/datasets/datasets/apple2orange/test")
#apple2orangeデータセットを用いる datasetは学習用,dataset_testは表示用に用いる
```

```python
#https://lp-tech.net/articles/hzfn7
#GeneratorはU-Netを使用
#の図のU-Netを参考に実装
#upsampleにconvtranspose,instance normalizeをconvの後に入れている

#画像サイズをswtride2のconvで半分にした後2回conv
class DownSamp(nn.Module):
    def __init__(self, in_ch, out_ch):
        super(DownSamp, self).__init__()
        self.conv1 = nn.Conv2d(in_ch, out_ch, kernel_size=3, stride=2,padding=1)
        self.conv2 = nn.Conv2d(out_ch, out_ch, kernel_size=3, stride=1,padding=1)
        self.conv3 = nn.Conv2d(out_ch, out_ch, kernel_size=3, stride=1,padding=1)
    def forward(self, x):
        x = self.conv1(x)
        x = F.instance_norm(x)
        x = F.leaky_relu(x)
        x = self.conv2(x)
        x = F.instance_norm(x)
        x = F.leaky_relu(x)
        x = self.conv3(x)
        x = F.instance_norm(x)
        x = F.leaky_relu(x)
        return x

    
#画像サイズをswtride2のconvtransposeで2倍にした後catして2回conv
class UpSamp(nn.Module):
    def __init__(self, in_ch, cat_ch, out_ch):
        super(UpSamp, self).__init__()
        self.deconv = nn.ConvTranspose2d(in_ch,in_ch//2,kernel_size=4, stride=2,padding=1)
        #forwardでcatする catするサイズはアップサンプル後
        self.conv1 = nn.Conv2d(in_ch//2+cat_ch, out_ch, kernel_size=3, stride=1,padding=1)
        self.conv2 = nn.Conv2d(out_ch, out_ch, kernel_size=3, stride=1,padding=1)
    def forward(self,x, cat):
        x = self.deconv(x)
        x = F.instance_norm(x)
        x = F.leaky_relu(x)
        x = torch.cat((x, cat),axis=1)
        x = self.conv1(x)
        x = F.instance_norm(x)
        x = F.leaky_relu(x)
        x = self.conv2(x)
        x = F.instance_norm(x)
        x = F.leaky_relu(x)
        return x
d = DownSamp(32,64)
u = UpSamp(64,32,32)
d(torch.zeros((1,32,64,64)))
u(torch.zeros((1,64,32,32)),torch.zeros((1,32,64,64))).shape

#最初の層と最後の層のみkernel_size=7
class UNet(nn.Module):
    def __init__(self):
        super(UNet,self).__init__()
        self.in1 = nn.Conv2d(3, 32, kernel_size=7,stride=1,padding=3)
        self.in2 = nn.Conv2d(32, 32, kernel_size=3,stride=1,padding=1)
        self.down1 = DownSamp(32,64)
        self.down2 = DownSamp(64,128)
        self.down3 = DownSamp(128,256)
        self.down4 = DownSamp(256,512)
        self.up4 = UpSamp(512,256,256)
        self.up3 = UpSamp(256,128,128)
        self.up2 = UpSamp(128,64,64)
        self.up1 = UpSamp(64,32,32)
        self.out = nn.Conv2d(32,3, kernel_size=7, stride=1, padding=3)
    def forward(self,x):
        x = self.in1(x)
        x = F.instance_norm(x)
        x = F.leaky_relu(x)
        x1 = self.in2(x)
        x1 = F.instance_norm(x1)
        x1 = F.leaky_relu(x)
        x2 = self.down1(x1)
        x3 = self.down2(x2)
        x4 = self.down3(x3)
        x5 = self.down4(x4)
        x = self.up4(x5,x4)
        x = self.up3(x, x3)
        x = self.up2(x, x2)
        x = self.up1(x, x1)
        x = self.out(x)
        x = torch.tanh(x)
        return x
g = UNet().to(device)
g(torch.zeros((1,3,256,256)).to(device)).shape
```

```python
from torch.nn.utils.spectral_norm import spectral_norm
#spectral normを入れたpatch gan discriminator
#最終的に(batch_size,64*64)を出力
#Mean Squared lossを用いて学習
class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator, self).__init__()        
        self.main = nn.Sequential(
            spectral_norm(nn.Conv2d(3, 32, stride=1, kernel_size=3, padding=1)),
            nn.LeakyReLU(0.1),
            spectral_norm(nn.Conv2d(32, 64, stride=2, kernel_size=3, padding=1)),
            nn.LeakyReLU(0.1),
            spectral_norm(nn.Conv2d(64, 128, stride=1, kernel_size=3, padding=1)),
            nn.LeakyReLU(0.1),
            spectral_norm(nn.Conv2d(128, 256, stride=2, kernel_size=3, padding=1)),
            nn.LeakyReLU(0.1),
            spectral_norm(nn.Conv2d(256,512,stride=1,kernel_size=3,padding=1)),
            nn.LeakyReLU(0.1),
            #spectral_norm(nn.Conv2d(256,512,stride=2,kernel_size=3,padding=1)),
            #nn.LeakyReLU(0.1),
            #nn.Conv2d(1024,1024,stride=1,kernel_size=1,padding=0),
            spectral_norm(nn.Conv2d(512,1,stride=1,kernel_size=1,padding=0)),
            nn.Flatten(),
        )
    def forward(self, x):
        return (self.main(x).squeeze_(1))
d = Discriminator()
d(torch.zeros((1,3,256,256))).shape
```

```python
#メディア実験のプロット関数を少し書き換え
%matplotlib inline
import matplotlib.pyplot as plt
from IPython import display

def sample_images(gen_ab,gen_ba,dataset):
    #os.makedirs('images/%s' % self.dataset_name, exist_ok=True)
    r, c = 4, 5
    dataloader = torch.utils.data.DataLoader(dataset, batch_size=5,shuffle=False)
    #データセットから最初の5枚のみ使う
    imgs_A, imgs_B = next(iter(dataloader))
    fake_A = gen_ba(imgs_B.to(device)).detach()
    fake_B = gen_ab(imgs_A.to(device)).detach()
    gen_imgs = np.concatenate([
        imgs_A.cpu().numpy().transpose(0,2,3,1),
        fake_B.cpu().numpy().transpose(0,2,3,1),
        imgs_B.cpu().numpy().transpose(0,2,3,1), 
        fake_A.cpu().numpy().transpose(0,2,3,1),
    ])

    # Rescale images 0 - 1
    gen_imgs = 0.5 * gen_imgs + 0.5

    titles = ['Input_A', 'Generated_B', 'Input_B', 'Generated_A']
    fig, axs = plt.subplots(r, c, figsize=(10,6))
    cnt = 0
    for i in range(r):
        for j in range(c):
            axs[i,j].imshow(gen_imgs[cnt])
            axs[i,j].set_title(titles[i])
            axs[i,j].axis('off')
            cnt += 1
    
    display.clear_output(wait = True)
    display.display(fig)
    plt.close()
```

```python
batch_size=4
epoch = 30
dataloader = torch.utils.data.DataLoader(dataset, batch_size=batch_size,shuffle=True)
g_ab = UNet()
g_ba = UNet()
d_a = Discriminator()
d_b = Discriminator()
g_ab = nn.DataParallel(g_ab)
g_ba = nn.DataParallel(g_ba)
d_a = nn.DataParallel(d_a)
d_b = nn.DataParallel(d_b)
#データ並列化
g_ab = g_ab.to(device)
g_ba = g_ba.to(device)
d_a = d_a.to(device)
d_b = d_b.to(device)
#zero_ = torch.zeros((batch_size,64)).to(device)
#one_ = torch.ones((batch_size,64)).to(device)
lossfunc = nn.MSELoss()
opt_g_ab = opt.SGD(params=g_ab.parameters(),lr=1e-3, momentum=0.8)
opt_g_ba = opt.SGD(params=g_ba.parameters(),lr=1e-3, momentum=0.8)
opt_d_a = opt.SGD(params=d_a.parameters(),lr=1e-2, momentum=0.8)
opt_d_b = opt.SGD(params=d_b.parameters(),lr=1e-2, momentum=0.8)
#g_opt = opt.Adam(params=g.parameters())
#d_opt = opt.Adam(params=d.parameters())
itr = 0
for ep in range(epoch):
    for (img_a, img_b) in dataloader:
        itr+=1
        img_a = img_a.to(device)
        img_b = img_b.to(device)
        fake_b = g_ab(img_a)
        fake_a = g_ba(img_b)
        #A,Bそれぞれを相互に変換
        
        rec_a = g_ba(fake_b)
        rec_b = g_ab(fake_a)
        #元のドメインに変換しなおす
        
        id_a = g_ba(img_a)
        id_b = g_ab(img_b)
        #変換元ドメイン以外の画像を入れても変化しないようにする
        
        zero_ = torch.zeros((img_a.shape[0],4096)).to(device)
        one_ = torch.ones((img_a.shape[0],4096)).to(device)
        
        l_g_aba_rec = torch.mean(torch.abs(img_a-rec_a))
        l_g_bab_rec = torch.mean(torch.abs(img_b-rec_b))
        #変換して元のドメインに戻したときに元の画像が復元できるようにする imgAとg_ab(g_ba(imgA))のL1 lossと反対も
        
        l_g_ab_adv = lossfunc(d_a(fake_a),one_)
        l_g_ba_adv = lossfunc(d_b(fake_b),one_)
        #変換された画像が変換先ドメインの画像と見分けがつかないようにする LSGAN loss
        
        l_g_ab_id = torch.mean(torch.abs(img_b-id_b))
        l_g_ba_id = torch.mean(torch.abs(img_a-id_a))
        #変換ドメイン以外の画像を入れても画像が変化しないようにする imgAとg_ba(imgA)のL1 と反対も
        
        l_g = (l_g_ab_adv + l_g_ba_adv) + (l_g_aba_rec+l_g_bab_rec)*0.1 + (l_g_ab_id + l_g_ba_id)*0.025
        #generator lossのsum
        
        g_ab.zero_grad()
        g_ba.zero_grad()
        l_g.backward()
        opt_g_ab.step()
        opt_g_ba.step()
        #generatorの学習
        
        l_d_a_adv = lossfunc(d_a(fake_a.detach()),zero_)+lossfunc(d_a(img_a),one_)
        l_d_b_adv = lossfunc(d_b(fake_b.detach()),zero_)+lossfunc(d_b(img_b),one_)
        #本物と変換された画像の見分けがつくようにする LSGAN loss
        
        l_d = l_d_a_adv + l_d_b_adv
        
        d_a.zero_grad()
        d_b.zero_grad()
        l_d.backward()
        opt_d_a.step()
        opt_d_b.step()
        #discriminatorの学習
                

    if ep % 1 == 0:
        sample_images(g_ab,g_ba,dataset_test)
        print("epoch: %d"%ep)
        print("generator a->b adversarial loss: %f"%l_g_ab_adv)
        print("generator a->b->a reconstruct loss: %f"%l_g_aba_rec)
        print("generator a->b identity mapping loss: %f"%l_g_ab_id)
        print("generator b->a adversarial loss: %f"%l_g_ba_adv)
        print("generator b->a->b reconstruct loss: %f"%l_g_bab_rec)
        print("generator b->a identity mapping loss: %f"%l_g_ba_id)
        print("discriminator a advarsarial loss: %f"%l_d_a_adv)
        print("discriminator b advarsarial loss: %f"%l_d_b_adv)
```

最初identity mapping lossを実装しなかったところ全体の色が変化してしまったのでidentity mapping lossによって構造を保つことができるのかなと感じた。
generatorの最終層でtanhを用いたがsigmoidとの違いがよく違いが判らなかった。データセットを見ると大きさや写り方が様々な画像があったのでもう少しきれいなデータセットならよりうまく変換できると考える。
